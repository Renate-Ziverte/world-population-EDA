## An example: visualizing data in the notebook ✨

Below is an example of a code cell. We'll visualize some simple data using two popular packages in Python. We'll use [NumPy](https://numpy.org/) to create some random data, and [Matplotlib](https://matplotlib.org) to visualize it.

Note how the code and the results of running the code are bundled together.

```python
#import sll relevent libraries - please note, was not able to download seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#download the data set and display first 5 rows
df = pd.read_csv("world_population_data.csv")


df.head()
```

```python
#show column names
df.columns
```

```python
#gain an overview of the data set including data  types& missing values
df.info
```

```python
# Check data types of columns
print(df.dtypes)
```

```python
df.describe
```

```python
#display the count of countries per continent in this data set
value_counts = df['continent'].value_counts()
print(value_counts)
```

```python
#check for missing data 
#it appears there are no missing values in this data set
df.isna().sum()
```

```python
# Create a bar plot for the distribution of countries across continents
value_counts = df['continent'].value_counts()

value_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Countries Across Continents')
plt.xlabel('Continent')
plt.ylabel('Number of Countries')
plt.show()
```

```python
# Pie chart for world population distribution by continent
population_by_continent = df.groupby('continent')['2023 population'].sum()

population_by_continent.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral', 'gold', 'lightskyblue'])
plt.title('World Population Distribution by Continent in 2023')
plt.show()
```

```python
# Histogram for growth rate distribution with Matplotlib
plt.figure(figsize=(8, 6))
plt.hist(df['growth rate'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Growth Rates')
plt.xlabel('Growth Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for top 10 countries by population in 2023
top_countries_2023 = df.nlargest(10, '2023 population')

top_countries_2023.plot(x='country', y='2023 population', kind='bar', color='purple')
plt.title('Top 10 Countries by Population in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
# Box plot for population density
df.boxplot(column='density (km²)', by='continent', color='skyblue')
plt.title('Population Density by Continent')
plt.ylabel('Density (km²)')
plt.xlabel('Continent')
plt.show()
```

```python
# Bar plot for top 10 countries by population in 2023 with Matplotlib
plt.figure(figsize=(10, 6))
plt.barh(top_countries_2023['country'], top_countries_2023['2023 population'], color='skyblue')
plt.title('Top 10 Countries by Population in 2023')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
#identify countries with highest population growth rates
# Convert 'growth rate' column to numeric
df['growth rate'] = pd.to_numeric(df['growth rate'].str.rstrip('%'), errors='coerce')

# Identify countries with the highest population growth rates
highest_growth_rates = df.nlargest(5, 'growth rate')

print("Countries with the Highest Population Growth Rates:")
print(highest_growth_rates[['country', 'growth rate']])
```

```python
#identify countries with lovest population growth rates
# Remove non-numeric characters from 'growth rate' and convert to numeric
df['growth rate'] = pd.to_numeric(df['growth rate'].replace('[^\d.]', '', regex=True), errors='coerce')

# Identify countries with the lowest population growth rates
lowest_birth_rates = df.nsmallest(5, 'growth rate')

print("Countries with the Lowest Birth Rates:")
print(lowest_birth_rates[['country', 'growth rate']])
```

```python
#Create a choropleth map to visualize world population distribution by country
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the world map GeoDataFrame
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world map with the population data
merged_data = world.merge(df, how='left', left_on='iso_a3', right_on='cca3')

# Plot the choropleth map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

merged_data.plot(column='2023 population', ax=ax, legend=True,
                 legend_kwds={'label': "Population by Country",
                              'orientation': "horizontal"})

plt.title('World Population Distribution by Country (2023)')
plt.show()
```

```python


import matplotlib.pyplot as plt

# Specify the country for analysis
selected_country = 'Latvia'

# Extract relevant columns for the selected country
projections_columns = ['country', '2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population']
selected_country_projections = df[df['country'] == selected_country][projections_columns]

# Set 'country' as the index for easier plotting
selected_country_projections.set_index('country', inplace=True)

# Plot line chart for population projections over time for the selected country
plt.figure(figsize=(15, 8))
selected_country_projections.T.plot(marker='o', figsize=(15, 8))
plt.title(f'Population Projections Over Time for {selected_country}')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(title='Year', bbox_to_anchor=(1, 1))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Specify the country for analysis
selected_country = 'India'

# Extract relevant columns for the selected country
projections_columns = ['country', '2023 population', '2022 population', '2020 population', '2015 population', '2010 population', '2000 population', '1990 population', '1980 population', '1970 population', 'growth rate']
selected_country_data = df[df['country'] == selected_country][projections_columns]

# Plot line chart for the growth rate over time for the selected country
plt.figure(figsize=(12, 8))
plt.plot(selected_country_data.columns[4:-1], selected_country_data.iloc[0, 4:-1], marker='o')
plt.title(f'Population Growth Rate Over Time for {selected_country}')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.show()
```
